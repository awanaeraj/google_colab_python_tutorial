```python
students = [
    {"name": "Ali", "marks": 85},
    {"name": "Ahmed", "marks": 78},
    {"name": "Shoaib", "marks": 92},
    {"name": "Ayesha", "marks": 88},
    {"name": "Usman", "marks": 70},
    {"name": "Zain", "marks": 95},
    {"name": "Aeraj", "marks": 80},
    {"name": "Moiz", "marks": 76},
    {"name": "Fatima", "marks": 89},
    {"name": "Hamza", "marks": 82}
]


def calculate_average(students):
    total = sum(student["marks"] for student in students)
    return total / len(students)

def get_top_student(students):
    return max(students, key=lambda x: x["marks"])

def get_above_average(students, avg):
    return [s for s in students if s["marks"] > avg]




average = calculate_average(students)
top_student = get_top_student(students)
above_avg_students = get_above_average(students, average)

print("Class Average:", average)
print("Top Student:", top_student["name"], "-", top_student["marks"])

print("\nStudents Above Average:")
for s in above_avg_students:
    print(s["name"], "-", s["marks"])
```

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True


def is_palindrome(value):
    value_str = str(value)
    return value_str == value_str[::-1]


def fibonacci(n):
    sequence = []
    a, b = 0, 1

    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence


def factorial(n):
    if n < 0:
        return None

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result


while True:
    print("\n===== Algorithm Practice Menu =====")
    print("1. Prime Number Checker")
    print("2. Palindrome Checker")
    print("3. Fibonacci Sequence")
    print("4. Factorial Calculator")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    # Prime Number
    if choice == '1':
        num = int(input("Enter a number: "))
        if is_prime(num):
            print(f"{num} is a PRIME number")
        else:
            print(f"{num} is NOT a prime number")

    # Palindrome
    elif choice == '2':
        value = input("Enter a value (number or string): ")
        if is_palindrome(value):
            print(f"{value} is a PALINDROME")
        else:
            print(f"{value} is NOT a palindrome")

    # Fibonacci
    elif choice == '3':
        n = int(input("Enter number of terms: "))
        print("Fibonacci sequence:", fibonacci(n))

    # Factorial
    elif choice == '4':
        num = int(input("Enter a number: "))
        result = factorial(num)
        if result is None:
            print("Factorial not defined for negative numbers")
        else:
            print(f"Factorial of {num} is {result}")

    # Exit
    elif choice == '5':
        print("Exiting program... Goodbye!")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
import numpy as np

months = np.array([
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])

sales = np.array([1200, 1500, 1800, 1700, 2000, 2100,
                  1900, 2200, 2500, 2400, 2300, 2600])

total_sales = np.sum(sales)
print("Total Sales:", total_sales)

avg_sales = np.mean(sales)
print("Average Monthly Sales:", avg_sales)
max_index = np.argmax(sales)
min_index = np.argmin(sales)

print("Highest Sales Month:", months[max_index], "-", sales[max_index])
print("Lowest Sales Month:", months[min_index], "-", sales[min_index])

top_indices = np.argsort(sales)[-3:][::-1]

print("\nTop 3 Performing Months:")
for i in top_indices:
    print(months[i], "-", sales[i])
```

```python
import numpy as np

# Generate random scores between 0 and 100
np.random.seed(42)  # for consistent results
scores = np.random.randint(0, 101, 100)

print("Student Scores:\n", scores)

avg_score = np.mean(scores)
print("\nAverage Score:", avg_score)

pass_count = np.sum(scores >= 50)
fail_count = np.sum(scores < 50)

print("Pass Count:", pass_count)
print("Fail Count:", fail_count)

bins = [0, 50, 70, 90, 100]

distribution, _ = np.histogram(scores, bins)

print("\nScore Distribution:")
print("0-49  (Fail):", distribution[0])
print("50-69 (Average):", distribution[1])
print("70-89 (Good):", distribution[2])
print("90-100 (Excellent):", distribution[3])

top_indices = np.argsort(scores)[-10:][::-1]

top_scores = scores[top_indices]

print("\nTop 10 Performers Scores:")
print(top_scores)
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ali", "Ahmed", "Sara", "Ali", "Zain", "Ayesha", None, "Usman", "Hina", "Sara"],
    "Department": ["IT", "hr", "Finance", "IT", "IT", "marketing", "HR", "Finance", "Hr", "finance"],
    "Salary": [50000, 40000, np.nan, 50000, 55000, 45000, 42000, np.nan, 42000, 60000],
    "Experience": [2, 3, 5, 2, np.nan, 2, 3, 6, 3, 5]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

print("\nData Info:\n")
print(df.info())

print("\nMissing Values:\n")
print(df.isnull().sum())

df = df.dropna(subset=["Name"])

df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

df["Experience"] = df["Experience"].fillna(df["Experience"].median())

df = df.drop_duplicates()

df["Department"] = (
    df["Department"]
    .str.strip()
    .str.lower()
    .replace({
        "hr": "HR",
        "it": "IT",
        "finance": "Finance",
        "marketing": "Marketing"
    })
)

df = df.reset_index(drop=True)

print("\nCleaned Dataset:\n")
print(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

sns.set_style("whitegrid")

print("===== TITANIC DATA VISUALIZATION =====\n")


plt.figure()
sns.countplot(x='survived', data=df)
plt.title("Survival Distribution (0 = No, 1 = Yes)")
plt.show()

print("Observation: More passengers did NOT survive than survived.\n")


plt.figure()
sns.countplot(x='sex', data=df)
plt.title("Gender Distribution")
plt.show()

print("Observation: Male passengers are more than female passengers.\n")


plt.figure()
sns.countplot(x='pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()

print("Observation: Most passengers were in 3rd class.\n")


plt.figure()
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()

print("Observation: Most passengers were between 20–40 years old.\n")


plt.figure()
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()

print("Observation: 1st class passengers had higher survival chances.\n")
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')

print("\n================ TITANIC DATASET REPORT ================\n")


print("1. DATASET OVERVIEW")
print("---------------------")
print(f"Total Passengers: {len(df)}")
print(f"Total Features: {df.shape[1]}")
print("\n")


print("2. SURVIVAL TRENDS")
print("---------------------")

survival_counts = df['survived'].value_counts()
survival_rate = df['survived'].mean() * 100

print(f"Survived: {survival_counts[1]}")
print(f"Not Survived: {survival_counts[0]}")
print(f"Survival Rate: {survival_rate:.2f}%")

gender_survival = df.groupby('sex')['survived'].mean() * 100
print("\nSurvival by Gender (%):")
print(gender_survival)

class_survival = df.groupby('pclass')['survived'].mean() * 100
print("\nSurvival by Passenger Class (%):")
print(class_survival)

print("\n")


print("3. DEMOGRAPHIC OBSERVATIONS")
print("---------------------")

print(f"Male Passengers: {df['sex'].value_counts()['male']}")
print(f"Female Passengers: {df['sex'].value_counts()['female']}")

print("\nPassenger Class Distribution:")
print(df['pclass'].value_counts().sort_index())

print("\nAge Summary:")
print(df['age'].describe())

print("\n")


print("4. DATA QUALITY OBSERVATIONS")
print("---------------------")

missing = df.isnull().sum()
print("Missing Values per Column:")
print(missing)

print("\nKey Issues:")
print("- Age column has missing values")
print("- Cabin column has many missing values")
print("- Embarked column has few missing values")

print("\n")


print("5. POTENTIAL PREDICTIVE FEATURES")
print("---------------------")

print("- Sex: Strong impact on survival")
print("- Pclass: Higher class = higher survival chance")
print("- Age: Children had better survival rate")
print("- Fare: Higher fare linked to survival")
print("- SibSp/Parch: Family size affects survival")
print("- Embarked: Slight influence on survival")

print("\n========================================================")
print("END OF REPORT")
print("========================================================")
```
