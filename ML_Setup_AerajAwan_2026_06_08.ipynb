```python
students = [
    {"name": "Ali", "marks": 85},
    {"name": "Ahmed", "marks": 78},
    {"name": "Shoaib", "marks": 92},
    {"name": "Ayesha", "marks": 88},
    {"name": "Usman", "marks": 70},
    {"name": "Zain", "marks": 95},
    {"name": "Aeraj", "marks": 80},
    {"name": "Moiz", "marks": 76},
    {"name": "Fatima", "marks": 89},
    {"name": "Hamza", "marks": 82}
]


def calculate_average(students):
    total = sum(student["marks"] for student in students)
    return total / len(students)

def get_top_student(students):
    return max(students, key=lambda x: x["marks"])

def get_above_average(students, avg):
    return [s for s in students if s["marks"] > avg]




average = calculate_average(students)
top_student = get_top_student(students)
above_avg_students = get_above_average(students, average)

print("Class Average:", average)
print("Top Student:", top_student["name"], "-", top_student["marks"])

print("\nStudents Above Average:")
for s in above_avg_students:
    print(s["name"], "-", s["marks"])
```

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True


def is_palindrome(value):
    value_str = str(value)
    return value_str == value_str[::-1]


def fibonacci(n):
    sequence = []
    a, b = 0, 1

    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence


def factorial(n):
    if n < 0:
        return None

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result


while True:
    print("\n===== Algorithm Practice Menu =====")
    print("1. Prime Number Checker")
    print("2. Palindrome Checker")
    print("3. Fibonacci Sequence")
    print("4. Factorial Calculator")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    # Prime Number
    if choice == '1':
        num = int(input("Enter a number: "))
        if is_prime(num):
            print(f"{num} is a PRIME number")
        else:
            print(f"{num} is NOT a prime number")

    # Palindrome
    elif choice == '2':
        value = input("Enter a value (number or string): ")
        if is_palindrome(value):
            print(f"{value} is a PALINDROME")
        else:
            print(f"{value} is NOT a palindrome")

    # Fibonacci
    elif choice == '3':
        n = int(input("Enter number of terms: "))
        print("Fibonacci sequence:", fibonacci(n))

    # Factorial
    elif choice == '4':
        num = int(input("Enter a number: "))
        result = factorial(num)
        if result is None:
            print("Factorial not defined for negative numbers")
        else:
            print(f"Factorial of {num} is {result}")

    # Exit
    elif choice == '5':
        print("Exiting program... Goodbye!")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
import numpy as np

months = np.array([
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])

sales = np.array([1200, 1500, 1800, 1700, 2000, 2100,
                  1900, 2200, 2500, 2400, 2300, 2600])

total_sales = np.sum(sales)
print("Total Sales:", total_sales)

avg_sales = np.mean(sales)
print("Average Monthly Sales:", avg_sales)
max_index = np.argmax(sales)
min_index = np.argmin(sales)

print("Highest Sales Month:", months[max_index], "-", sales[max_index])
print("Lowest Sales Month:", months[min_index], "-", sales[min_index])

# Get indices of top 3 sales
top_indices = np.argsort(sales)[-3:][::-1]

print("\nTop 3 Performing Months:")
for i in top_indices:
    print(months[i], "-", sales[i])
```
